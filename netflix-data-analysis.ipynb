# STEP 3: PROCESS

We are using Python to prepare and process the data.

#### 3.1 Preparing the Environment

First, I need to **import libraries** (think of libraries as toolboxes — each one has special tools for specific tasks):

- `pandas` → for working with tables of data (like Excel in Python)
- `numpy` → for doing math calculations
- `matplotlib` → for making charts and graphs

```python
# 📦 Import the libraries we need
# Think of these as toolboxes — each has tools for different jobs

import pandas as pd          # for working with data tables
import numpy as np           # for math calculations
import matplotlib.pyplot as plt  # for making charts

print("✅ Libraries imported successfully!")
```

#### 3.2 Importing the Dataset

Now I load the CSV file into Python using `pd.read_csv()`. This turns the file into a **DataFrame** — which is basically a table with rows and columns, just like a spreadsheet.

```python
# 📂 Load the dataset from the CSV file
# Make sure netflix_titles.csv is in the same folder as this notebook!

df = pd.read_csv("netflix_titles.csv")

print("✅ Dataset loaded!")
print(f"Shape of the data: {df.shape}")
# .shape gives us (number of rows, number of columns)
```

#### 3.3 Data Cleaning and Exploration

##### Steps:

1. Observe and familiarize with the data
2. Check for null or missing values
3. Fix data types and formats
4. Create new useful columns

Previewing the first 10 rows using the `head()` function to understand what the data looks like.

```python
# 👀 Preview the first 10 rows of the data
# head(10) shows the first 10 rows — great for getting a quick look!

df.head(10)
```

Now let me check the basic information — number of rows/columns, column names, and data types.

```python
# 📋 Check basic information about the dataset
# .info() is like a summary card for your data!

df.info()
```

Checking for **missing values** — these are empty cells in our data. We need to know where they are before we can work with the data properly.

```python
# ❓ Check for missing (null) values in each column
# isnull() marks each cell as True/False (missing or not)
# .sum() counts how many True values (= how many missing) per column

missing_values = df.isnull().sum()
print("Missing values per column:")
print(missing_values)
```

From the above, we can see that columns like `director`, `cast`, and `country` have many missing values. That's okay — not every title has a known director or cast listed.

Now let me **fix the `date_added` column** — it currently looks like a text string (e.g., `"January 1, 2020"`) but we need Python to understand it as an actual date.

```python
# 📅 Convert 'date_added' from text to a real date format
# pd.to_datetime() tells Python: "hey, this column is a date, not just text!"

df["date_added"] = pd.to_datetime(df["date_added"].str.strip(), format="%B %d, %Y")

# Confirm the conversion worked
print("Data type of date_added after conversion:", df["date_added"].dtype)
df["date_added"].head(5)
```

Now I'll create **new columns** from `date_added` to make analysis easier:

- `year_added` — the year the title was added to Netflix
- `month_added` — the month it was added

```python
# 🗓️ Create new columns: year_added and month_added
# .dt.year extracts just the year from a date
# .dt.month extracts just the month

df["year_added"] = df["date_added"].dt.year
df["month_added"] = df["date_added"].dt.month

# Quick preview
df[["title", "date_added", "year_added", "month_added"]].head(5)
```

Let me also **clean the `country` column** by keeping only the first country listed (since some titles have multiple countries separated by commas).

```python
# 🌍 Keep only the FIRST country listed in the 'country' column
# Some rows look like: "United States, India, UK" — we just want "United States"
# .str.split(',') splits by comma, [0] gets the first part, .str.strip() removes extra spaces

df["country_main"] = df["country"].str.split(',').str[0].str.strip()

print("✅ New column 'country_main' created!")
df[["title", "country", "country_main"]].head(5)
```

Data cleaning is done! ✅ Let me take a final preview of the cleaned data.

---

```python
# ✅ Final preview of cleaned data
print(f"Dataset has {df.shape[0]} rows and {df.shape[1]} columns after cleaning.")
df.head()
```

# STEP 4: ANALYZE

#### 4.1 Perform Calculations

Now I'll use `describe()` to get basic statistics on numeric columns.

```python
# 📊 Get general statistics for all numeric columns
# describe() gives: count, mean, std, min, max, and percentiles

df.describe()
```

#### 4.2 Movies vs TV Shows

How many Movies and TV Shows are there on Netflix?

```python
# 🎬 Count how many Movies vs TV Shows
# value_counts() counts each unique value in a column

type_counts = df["type"].value_counts()
print("Content type breakdown:")
print(type_counts)

# Also calculate as percentage
print("\nPercentage:")
print(round(type_counts / type_counts.sum() * 100, 1))
```

#### 4.3 Top 10 Countries by Content

Which countries produce the most content on Netflix?

```python
# 🌍 Find the top 10 countries with most Netflix content
# dropna() removes rows where country is missing
# value_counts() counts titles per country
# head(10) keeps only the top 10

top_countries = df["country_main"].dropna().value_counts().head(10)
print("Top 10 Countries by number of titles:")
print(top_countries)
```

#### 4.4 Growth Over the Years

How many new titles did Netflix add each year?

```python
# 📈 Count how many titles were added per year
# dropna() removes rows where date_added was missing
# sort_index() sorts by year (oldest to newest)

titles_per_year = df.dropna(subset=["year_added"])["year_added"].value_counts().sort_index()
print("Titles added per year:")
print(titles_per_year)
```

#### 4.5 Most Common Ratings

What are the most common age ratings?

```python
# 🔞 Count content by age rating

rating_counts = df["rating"].value_counts().head(10)
print("Most common ratings:")
print(rating_counts)
```

#### 4.6 Top Genres

The `listed_in` column contains genres separated by commas (e.g., `"Dramas, International Movies"`). I need to split them up and count each one individually.

```python
# 🎭 Find the most common genres
# Each row in 'listed_in' can have multiple genres separated by commas
# We split them, flatten the list, and count each genre individually

# Step 1: Split genres and put all of them in one big list
all_genres = df["listed_in"].dropna().str.split(", ")   # split by comma+space

# Step 2: 'explode' turns [['A','B'],['C']] into separate rows: A, B, C
genre_series = all_genres.explode()

# Step 3: Count and show top 10
top_genres = genre_series.value_counts().head(10)
print("Top 10 Genres on Netflix:")
print(top_genres)
```

# STEP 5: SHARE

Now let's create **visualizations** to present our findings in a clear and visual way!

#### 5.1 Chart 1 — Movies vs TV Shows (Pie Chart)

```python
# 🥧 Pie chart: Movies vs TV Shows
import matplotlib.pyplot as plt

type_counts = df["type"].value_counts()

plt.style.use("default")
plt.figure(figsize=(6, 6))   # set chart size

plt.pie(
    type_counts,
    labels=type_counts.index,
    autopct="%1.1f%%",         # show percentage on chart
    colors=["#E50914", "#B81D24"],   # Netflix red shades
    startangle=90,
    wedgeprops={"edgecolor": "white", "linewidth": 2}
)

plt.title("Movies vs TV Shows on Netflix", fontsize=14, fontweight="bold")
plt.show()
```

##### **Finding 1: Movies dominate Netflix**

From the pie chart:

1. **Movies** make up about **70%** of all Netflix titles.
2. **TV Shows** make up the remaining **30%**.
3. This shows Netflix has heavily invested in movies, but TV Shows are growing — they attract long-term subscribers.

---

#### 5.2 Chart 2 — Top 10 Countries (Horizontal Bar Chart)

```python
# 🌍 Horizontal bar chart: Top 10 countries by number of titles
import matplotlib.pyplot as plt

top_countries = df["country_main"].dropna().value_counts().head(10)

plt.style.use("default")
plt.figure(figsize=(8, 5))   # wider chart

# barh = horizontal bar chart
# [::-1] reverses the order so the highest is at the top
plt.barh(
    top_countries.index[::-1],
    top_countries.values[::-1],
    color="lightskyblue",
    edgecolor="black"
)

plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.title("Top 10 Countries by Number of Netflix Titles", fontsize=13, fontweight="bold")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.tight_layout()    # prevents labels from being cut off
plt.show()
```

##### **Finding 2: The United States leads by a huge margin**

1. The **United States** has by far the most titles — over 2,000+.
2. **India** is a strong second, showing Netflix's big push into the Indian market.
3. Countries like the UK, Canada, and France also appear — reflecting Netflix's global content strategy.

---

#### 5.3 Chart 3 — Netflix Growth Over the Years (Line Chart)

```python
# 📈 Line chart: Number of titles added per year
import matplotlib.pyplot as plt

titles_per_year = df.dropna(subset=["year_added"])["year_added"].value_counts().sort_index()

# Filter to years 2010 and later (data before that is very sparse)
titles_per_year = titles_per_year[titles_per_year.index >= 2010]

plt.style.use("default")
plt.figure(figsize=(10, 5))

plt.plot(
    titles_per_year.index,
    titles_per_year.values,
    color="#E50914",      # Netflix red!
    linewidth=2.5,
    marker="o",           # adds dots at each data point
    markersize=7
)

plt.xlabel("Year")
plt.ylabel("Number of Titles Added")
plt.title("Netflix Growth: Titles Added Per Year (2010–2021)", fontsize=13, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.6)
plt.xticks(titles_per_year.index, rotation=45)   # rotate x labels for readability
plt.tight_layout()
plt.show()
```

##### **Finding 3: Netflix grew explosively between 2016–2019**

1. Netflix started adding significantly more content around **2015–2016**, which matches when they expanded globally.
2. **2019 was the peak year** for new titles added to the platform.
3. A **drop in 2020** is likely due to COVID-19 disrupting productions worldwide.

---

#### 5.4 Chart 4 — Content Ratings (Bar Chart)

```python
# 🔞 Bar chart: Top content ratings on Netflix
import matplotlib.pyplot as plt

# Remove unusual/wrong values in rating column
valid_ratings = df[df["rating"].notna() & ~df["rating"].str.contains("min", na=False)]
rating_counts = valid_ratings["rating"].value_counts().head(10)

plt.style.use("default")
plt.figure(figsize=(9, 5))

plt.bar(
    rating_counts.index,
    rating_counts.values,
    color="salmon",
    edgecolor="black",
    width=0.6
)

plt.xlabel("Rating")
plt.ylabel("Number of Titles")
plt.title("Most Common Content Ratings on Netflix", fontsize=13, fontweight="bold")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

##### **Finding 4: Netflix is mostly adult-oriented content**

1. **TV-MA** (Mature Audiences only) is the most common rating — this is for adults 17+.
2. **TV-14** and **TV-PG** are next — suggesting Netflix also serves family audiences.
3. Very few titles are rated **G** or **TV-Y** (for young children), which might be a gap in Netflix's content for kids.

---

#### 5.5 Chart 5 — Top 10 Genres (Horizontal Bar Chart)

```python
# 🎭 Horizontal bar chart: Most common genres on Netflix
import matplotlib.pyplot as plt

all_genres = df["listed_in"].dropna().str.split(", ").explode()
top_genres = all_genres.value_counts().head(10)

plt.style.use("default")
plt.figure(figsize=(9, 6))

plt.barh(
    top_genres.index[::-1],
    top_genres.values[::-1],
    color="mediumseagreen",
    edgecolor="black"
)

plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.title("Top 10 Most Common Genres on Netflix", fontsize=13, fontweight="bold")
plt.grid(axis="x", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

##### **Finding 5: International Dramas and Documentaries are big!**

1. **International Movies** and **Dramas** are the top genres — showing Netflix's heavy investment in non-English content.
2. **Documentaries** and **Stand-Up Comedy** are also popular — these are cheaper to produce but attract loyal audiences.
3. The presence of **International TV Shows** confirms that shows like Korean dramas (K-dramas) are a major part of Netflix.
